### Colum Processor : Weight

```python
def column_text_processor(df, column):
    """
    Cleans and extracts numeric values from a text-based column, converting the result to float.

    This function performs the following steps:
        - Converts all values in the column to string type.
        - Removes all non-numeric characters except digits, slashes, and periods.
        - Extracts the first valid numeric pattern (e.g., "72.5" from "Weight: 72.5 kg").
        - Converts the extracted value to a numeric type (float), coercing invalid entries to NaN.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the column to process.
        column (str): The name of the column to clean and convert.

    Returns:
        pd.DataFrame: A new DataFrame with the cleaned numeric column.
    """
    df[column] = (
        df[column]
        .astype(str)
        .str.replace(r'[^\d/.]', '', regex=True)
        .str.extract(r'(\d+\.?\d*)')[0]
    )
    df[column] = pd.to_numeric(df[column], errors='coerce')
    return df

# Example usage
data = column_text_processor(data, 'Weight (kg)')
```

### Clean BP

```python
def pre_clean_blood_pressure(df, column):
    """
    Pre-cleans a blood pressure column by extracting valid systolic/diastolic (e.g., '120/80')
    patterns from messy or inconsistent text entries.

    This function:
        - Attempts to match clean patterns (e.g., '120/80').
        - Falls back to extracting systolic/diastolic values from noisy text (e.g., 'BP: 120 over 80').
        - Returns None for invalid or unparseable entries.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the blood pressure column.
        column (str): The name of the column to process.

    Returns:
        pd.DataFrame: The updated DataFrame with cleaned blood pressure values in 'SBP/DBP' format.
    """
    def extract_bp(value):
        value = str(value).strip()

        # Try matching clean format first (e.g., '120/80')
        match = re.match(r'^(\d{2,3})/(\d{2,3})$', value)
        if match:
            return f"{match.group(1)}/{match.group(2)}"

        # Try extracting from within messy text (e.g., 'BP: 120 over 80')
        match = re.search(r'(\d{2,3})\s*/\s*(\d{2,3})', value)
        if match:
            return f"{match.group(1)}/{match.group(2)}"

        return None  # Invalid or unrecognized format

    df[column] = df[column].apply(extract_bp)
    return df
```

```python
def clean_blood_pressure(df, column_name):
    """
    Cleans a blood pressure column in a DataFrame.

    - Removes trailing slashes
    - Handles missing or invalid values
    - Splits into Systolic (SBP) and Diastolic (DBP) columns
    - Converts to numeric format
    - Keeps the original column

    Args:
    - df (pd.DataFrame): The DataFrame containing the blood pressure column.
    - column_name (str): The column name of blood pressure readings.

    Returns:
    - pd.DataFrame: Updated DataFrame with 'SBP' and 'DBP' columns, keeping the original column.
    """
    # Ensure the column is a string type before processing
    df[column_name] = df[column_name].astype(str)

    # Remove trailing slashes
    df[column_name] = df[column_name].str.rstrip('/')

    # Replace standalone '/' or empty spaces with NaN
    df[column_name] = df[column_name].replace({'/': np.nan, '': np.nan})

    # Split into Systolic (SBP) and Diastolic (DBP) columns
    bp_df = df[column_name].str.split('/', expand=True)
    bp_df.columns = ['SBP', 'DBP']

    # Convert to numeric (invalid values become NaN)
    df['SBP'] = pd.to_numeric(bp_df['SBP'], errors='coerce')
    df['DBP'] = pd.to_numeric(bp_df['DBP'], errors='coerce')

    return df  # No dropping of the original column

data = clean_blood_pressure(data, 'Blood Pressure (mm/Hg)')
```

```python
age_mapping = {
    1: (20, 29),
    2: (30, 39),
    3: (40, 49),
    4: (50, 59),
    5: (60, 69),
    6: (70, 100)
}

def map_age_to_mean(df, column):
    """
    Maps categorical age group labels in the specified DataFrame column to their
    corresponding mean age values based on a predefined age range mapping.

    This function uses a helper function to compute the mean of each age range.
    If a value is not found in the mapping, it returns None to handle unexpected
    entries gracefully.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the age group column.
        column (str): The name of the column to be mapped.

    Returns:
        pd.Series: A Series containing the mean age values corresponding to each row.
    """
    def calculate_mean(value):
        if value in age_mapping:
            age_range = age_mapping[value]
            return sum(age_range) / 2
        return None  # Handle unexpected values gracefully


    df[column] = df[column].apply(calculate_mean)
    return df

data = map_age_to_mean(data, 'Age of Patient')
data['Age of Patient'].head()
```

```python
minerals_columns = ['Creatinine (Umol/L)', 'Sodium (mmol/L)', 'Potassium (mmol/L)',
                    'Glucose (mmol/L)', 'Urea (mmol/L)']
```

```python
def process_mineral_column(df, column):
    """
    Standardizes values in a mineral measurement column by converting all entries
    to mmol/L. It handles inputs in 'mg/dL', 'mmol/L', or plain numeric form,
    and gracefully returns None for invalid or unparseable entries.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the mineral data.
        column (str): The name of the column to process.

    Returns:
        pd.DataFrame: The updated DataFrame with standardized values in mmol/L.
    """
    def convert_value(value):
        try:
            if isinstance(value, str):
                value = value.strip().lower()
                if 'mg/dl' in value:
                    numeric_value = float(value.replace('mg/dl', '').strip())
                    return numeric_value * 0.0555  # Convert mg/dL to mmol/L
                elif 'mmol/l' in value:
                    return float(value.replace('mmol/l', '').strip())
                else:
                    return float(value)
            elif isinstance(value, (int, float)):
                return float(value)
        except (ValueError, TypeError):
            return None  # Handle invalid entries gracefully
        return None

    df[column] = df[column].apply(convert_value)
    return df

# Apply the transformation to all relevant columns
for col in minerals_columns:
    data_ = process_mineral_column(data_, col)
```

```python
def estimate_ckd_stage(row):
    """
    Estimates the stage of Chronic Kidney Disease (CKD) based on heuristic thresholds
    of Creatinine and Urea levels, and optionally Systolic/Diastolic Blood Pressure (SBP/DBP).

    The stages are determined using clinical-like logic with simplified thresholds:
        - Stage 5: Creatinine ≥ 13.85 µmol/L and Urea ≥ 140 mmol/L
        - Stage 4: Urea between 70–100 mmol/L and Creatinine between 2.5–3.5 µmol/L
        - Stage 3: Urea between 40–100 mmol/L and Creatinine between 1.8–3.4 µmol/L
        - Stage 2: Creatinine between 2.0–13.5 µmol/L and Urea between 70–150 mmol/L
        - Stage 1: Creatinine < 2.0 µmol/L and Urea < 30 mmol/L

    Parameters:
        row (pd.Series): A row from a DataFrame containing required clinical lab values.

    Returns:
        float or None: CKD stage (1.0–5.0) if matched; otherwise None.
    """
    cr = row['Creatinine (Umol/L)']
    urea = row['Urea (mmol/L)']
    sbp = row.get('SBP', None)
    dbp = row.get('DBP', None)

    if cr >= 13.85 and urea >= 140:
        return 5.0
    elif 70 <= urea <= 100 and 2.5 <= cr <= 3.5:
        return 4.0
    elif 40 <= urea <= 100 and 1.8 <= cr <= 3.4:
        return 3.0
    elif 2.0 <= cr <= 13.5 and 70 <= urea <= 150:
        return 2.0
    elif cr < 2.0 and urea < 30:
        return 1.0

    return None  # Unclassified or borderline case

# Apply the function to the DataFrame
data_['Stage Estimate Heuristic'] = data_.apply(estimate_ckd_stage, axis=1)
```

```python
def unify_stage(row):
    """
    Consolidates multiple CKD stage sources into a single unified stage value for each patient.

    The priority of sources is as follows:
        1. Clinically diagnosed stage ('Actual Diagnosed stage of the Acute and Chronic Kidney Disease')
        2. Heuristic estimate ('Stage Estimate Heuristic')
        3. Model-based estimate ('Stage Estimate Model')

    If none of these sources provide a valid stage, returns NaN.

    Parameters:
        row (pd.Series): A row from the DataFrame containing the different stage values.

    Returns:
        float or np.nan: The most reliable available CKD stage.
    """
    diagnosed_stage = row['Actual Diagnosed stage of the Acute and Chronic Kidney Disease']
    heuristic_stage = row['Stage Estimate Heuristic']
    model_stage = row['Stage Estimate Model']

    if not pd.isna(diagnosed_stage):
        return diagnosed_stage
    elif not pd.isna(heuristic_stage):
        return heuristic_stage
    elif not pd.isna(model_stage):
        return model_stage
    else:
        return np.nan

# Apply the function to the dataset
data_['Unified Stage'] = data_.apply(unify_stage, axis=1)
```

```python
def stage_based_impute(data_, features, target_col='Unified Stage'):
    """
    Performs imputation of missing values in the specified features based on CKD stage groups.

    For each feature:
        - If the feature is categorical (object type or few unique values), missing values are
          imputed using the mode within each CKD stage. If still missing, the global mode is used.
        - If the feature is numerical, missing values are imputed using the median within each CKD stage.
          If still missing, the global median is used.
        - A flag column (`<feature>_imputed`) is created to indicate which values were imputed.

    Parameters:
        data_ (pd.DataFrame): The input DataFrame containing clinical data.
        features (list of str): List of feature names to impute.
        target_col (str): The column name representing CKD stages (default: 'Unified Stage').

    Returns:
        pd.DataFrame: A new DataFrame with imputed values and corresponding flags.
    """
    df_filled = data_.copy()

    for feature in features:
        flag_col = f"{feature}_imputed"
        df_filled[flag_col] = df_filled[feature].isnull()

        is_categorical = df_filled[feature].dtype == 'object' or df_filled[feature].nunique(dropna=True) < 10

        if is_categorical:
            # Impute using mode per stage
            stage_modes = df_filled.groupby(target_col)[feature].agg(
                lambda x: x.mode().iloc[0] if not x.mode().empty else None
            )
            for stage, mode_val in stage_modes.items():
                mask = (df_filled[target_col] == stage) & (df_filled[feature].isnull())
                df_filled.loc[mask, feature] = mode_val

            # Fallback: Global mode
            if df_filled[feature].isnull().any():
                global_mode = df_filled[feature].mode().iloc[0]
                df_filled[feature].fillna(global_mode, inplace=True)

        else:
            # Impute using median per stage
            stage_medians = df_filled.groupby(target_col)[feature].median()
            for stage, median_val in stage_medians.items():
                mask = (df_filled[target_col] == stage) & (df_filled[feature].isnull())
                df_filled.loc[mask, feature] = median_val

            # Fallback: Global median
            if df_filled[feature].isnull().any():
                global_median = df_filled[feature].median()
                df_filled[feature].fillna(global_median, inplace=True)

    return df_filled
```

```python
def custom_oversample(df, target_col='Target', samples_per_class=100, random_state=42):
    """
    Balances a dataset by ensuring an equal number of samples for each class in the target column.

    For each class:
        - If there are more than `samples_per_class`, it performs undersampling.
        - If there are fewer, it performs custom oversampling by randomly recombining feature values.
        - If the class is already balanced, it is retained as is.

    Parameters:
        df (pd.DataFrame): Input DataFrame to balance.
        target_col (str): The name of the target column containing class labels.
        samples_per_class (int): Desired number of samples per class.
        random_state (int): Random seed for reproducibility.

    Returns:
        pd.DataFrame: A balanced DataFrame with equal number of samples per class.
    """
    np.random.seed(random_state)
    balanced = []

    for cls in df[target_col].unique():
        class_df = df[df[target_col] == cls].copy()
        n = len(class_df)

        if n > samples_per_class:
            # Undersample to target size
            sampled_df = class_df.sample(n=samples_per_class, random_state=random_state)

        elif n < samples_per_class:
            # Oversample by generating synthetic rows through random recombination
            features = class_df.drop(columns=[target_col])
            synthetic = []

            while len(synthetic) < (samples_per_class - n):
                new_row = {col: np.random.choice(features[col].values) for col in features.columns}
                synthetic.append(new_row)

            synthetic_df = pd.DataFrame(synthetic)
            synthetic_df[target_col] = cls
            sampled_df = pd.concat([class_df, synthetic_df], ignore_index=True)

        else:
            # Already balanced
            sampled_df = class_df

        balanced.append(sampled_df)

    # Shuffle the final DataFrame
    return pd.concat(balanced, ignore_index=True).sample(frac=1, random_state=random_state).reset_index(drop=True)
```
